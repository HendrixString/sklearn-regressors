```python
%load_ext autoreload
%autoreload 2

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from train import train_baseline_dummy, train_lasso, train_random_forest, train_NNLS
from eval import evaluate_estimator
```

```python
def train_and_eval():
    train_evals = [
        train_NNLS, 
        train_lasso,
        train_random_forest,
        train_baseline_dummy
    ]

    for t in train_evals:
        print(f"{t.__name__} regressor")
        m, x_test, y_test = t()

        # print metrics
        metrics = evaluate_estimator(m, x_test, y_test)
        print(metrics, '\n')

        # plot predictions
        Y_pred_test = m.predict(x_test) #predictions on training data
        plt.scatter(y_test, Y_pred_test)
        plt.xlabel("Actual Grade: $Y_i$")
        plt.ylabel("Predicted Grade: $\hat{Y}_i$")
        plt.title("Predicted Grade vs Actual Grade: $Y_i$ vs $\hat{Y}_i$")
        plt.show()

```

```python
train_and_eval()
```
